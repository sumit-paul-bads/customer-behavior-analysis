## **Category Mapping is needed as its mapped wrong**

```python
correct_mapping = {
    'Accessories' : ['Sunglasses' , 'Gloves' , 'Jewelry' , 'Hat' , 'Backpack' , 'Belt' , 'Bag' , 'Scarf' , 'Watch' , 'Socks' , 'Handbag'],
    'Footwear' : ['Boots' , 'Shoes' , 'Sandals' , 'Slippers' , 'Sneakers'],
    'Electronics' : ['Laptop' , 'Phone' , 'Headphones'],
    'Clothing' : ['T-shirt' , 'Shirt' , 'Shorts' , 'Hoodie' , 'Pants' , 'Jeans' , 'Skirt' , 'Sweater' , 'Dress' , 'Blouse'],
    'Outerwear' : ['Coat' , 'Jacket']
}
```

```python
# Create a flat mapping from item to category
item_to_category_map = {}
for category, items in correct_mapping.items():
    for item in items:
      item_to_category_map[item] = category

# Apply the mapping to the 'Category' column
df['Category'] = df['Item Purchased'].map(item_to_category_map)
```

## **Now Treating Null in Clothing**

```python
df.loc [ (df['Category'] == 'Clothing') & (df['Size'].isnull()) , 'Size'] = df.loc [df['Category'] == 'Clothing' , 'Size'].mode()[0]
```

```python
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].fillna (df.groupby('Item Purchased')['Purchase Amount (USD)'].transform('mean'))
```

```python
df['Review Rating'] = df['Review Rating'].fillna (df.groupby('Item Purchased')['Review Rating'].transform('mean'))

df['Review Rating'] = df['Review Rating'].round ()
```

```python
# Add a temporary column for the original index to preserve order within Customer ID groups
df_sorted = df.copy()
df_sorted['_original_index'] = df_sorted.index
df_sorted = df_sorted.sort_values(by=['Customer ID', '_original_index']).drop(columns=['_original_index'])

# Calculate the 'Purchase Amount (USD)' from the previous transaction for each customer.
# This will result in NaN for the first purchase of each customer within their group.
previous_purchase_amounts_shifted = df_sorted.groupby('Customer ID')['Purchase Amount (USD)'].shift(1)

# Align the shifted values back to the original DataFrame's index
previous_purchase_amounts_aligned = previous_purchase_amounts_shifted.reindex(df.index)

# Fill the NaN values in 'Previous Purchases' with these calculated amounts.
# Only fill where 'Previous Purchases' was originally NaN.
df['Previous Purchases'] = df['Previous Purchases'].fillna(previous_purchase_amounts_aligned)

# Fill any remaining NaN values (which are for the very first purchase of each customer)
# with 0, as per the requirement.
df['Previous Purchases'] = df['Previous Purchases'].fillna(0)
```
